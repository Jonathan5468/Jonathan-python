**Transferencia de calor en estado transitorio**

**Resistencia interna despreciable**
La transferencia de calor en estado transferencia la temperatura varia con el tiempo y posición, en esta parte la parcial de la temperatura no es igual a cero por la variación ya mencionada. 
El proceso es iniciado en cualquier momento cuando el sistema experimente un cambio en condiciones de operación y ocurre hasta alcanzar un nuevo estado estacionario o equilibrio térmico, es inducido por muchos cambios como en:
* Condiciones convectivas de superficie (h,∞)
* Condiciones radiativas de superficie (hf, Tsur)
* Temperatura en la superficie, se genera un flujo de calor 
* Generación interna de energía (no significa una espontaneidad sino una transformación de la energía)

**Ejemplo:** reacción química 
Los problemas pueden ser resueltos a través de tres métodos de solución:
* Resistencia interna despreciable 
* Soluciones exactas 
* Método por diferencias finitas 
Ecuación para la resistencia interna despreciable

$$ \frac{T-T_{\infty }}{T_{i}-T_{\infty }}= exp(-\frac{hA}{mCp}t) $$ 

* Los métodos analíticos son aplicados pero tienen muchos limitantes, por lo que tenemos que realizar estas ecuaciones

$$ T(x,y)=\tfrac{2T_{3}}{\pi}\sum_{n=1}^{\infty}\left [ \frac{(-1)^{n+1}+1}{n} \right ]\frac{sin\lambda _{n}x_{sinh}\lambda_{n}y}{sin\lambda_{n}W} $$

Esta ecuación por más que agreguemos el mayor número de términos siempre vamos incurrir a una aproximación, por lo que recurrimos a los métodos numéricos.

$$ T_{m}^{p+1}=F_{o}(T_{m-1}^{p}+T_{m+1}^{p})+(1-2F_{o})T_{p}^{m} $$

Dependiendo de la geometría del material usaremos las ecuaciones adecuadas sea la de coordenadas cilíndricas, esféricas y rectangulares. 
En el estado transitorio se asume que la distribución espacialmente uniforme
En un análisis de resistencia interna despreciables, en un caso general, se considera la convección, radiación y un flujo de calor aplicado a las superficies, finalmente también se supone la generación interna de energía en forma de calor.
**Resistencia interna del balance** 

Se realiza el balance de energía en donde

$$ Acumulación = Ingreso – Egreso + Generación + Consumo $$ 

$$ \frac{dE}{dt}=Ei - E_{e}+E_{g}-E_{c} ; En donde E_{c} es igual a cero $$
$$ \rho VCp\frac{dT}{dt}=q_{s}A_{s}-hA_{s}(T-T_{\infty })-E_{r}-E_{g} $$

En esta última ecuación se asume que la energía de egreso es debida a la convección y radiación, y que la energía de ingreso se debe a un flujo de calor qs.
En esta ecuación posterior la perdida de energía se encuentra a la izquierda del igual, y la derecha del igual se encuentra la Energía por convección (Newton).

$$ mCp\Delta T=h-A_{\infty }(T-T_{i}) $$

Se presenta el caso con una solución exacta T(0)=Ti

$$ \phi = T-T_{\infty} $$
$$ a= \frac{hA}{\rho Vc} $$
$$ b=\tfrac{\dot{q}Ah}{\rho Vc}$$
$$ {\phi }'= \phi-\frac{b}{a} $$

La ecuación tomando el balance de energía y la variación se utiliza la ecuación posterior 

$$ \frac{dT}{dt}=-\frac{hA}{mCp}(T-T_{\infty}) $$
La ecuación anterior es la ecuación de gobierno que describe las variaciones de Temperatura en función del tiempo para sistemas de resistencia interna despreciable o sistemas concentrados. Al ser una ecuación diferencial de primer orden, se requiere de una condición inicial para resolverla, luego debemos realizar los cambios de variables con las ecuaciones de solución exacta y obtenemos 

$$ \int_{\phi_{j}}^{\phi}\frac{d\phi }{dt}=\int_{0}^{1}-\frac{hA}{mCp}\phi $$

$$ \int_{\phi_{j}}^{\phi}\frac{d\phi }{dt}=-\frac{hA}{mCp}\int_{0}^{t}dt $$

$$ ln\frac{\phi }{\phi _{i}}=-\frac{hA}{mCp}t $$
$$ \frac{\phi }{\phi _{i}}=e^{-\frac{hA}{mCp}t} $$
Con esto se obtiene ña ecuación de temperatura en función del tiempo para sistemas concentrados o resistencia interna despreciable 

$$ \frac{T-T_{\infty }}{T_{i}-T_{\infty }}= exp(-\frac{hA}{mCp}t) $$ 

La ecuación muestra que toma un tiempo indefinido en alcanzar estado estacionario. A manera que la temperatura del sólido se acerca a la del fluido, la razón de transferencia de calor decrece haciendo imposible que el sólido alcance la temperatura del fluido.


### Introducción a CFD ###
Permite desarrollar la dinámica de fluidos de una forma computacional 
Esta dinámica es la aplicación de métodos numéricos para resolver problemas que involucren el flujo de fluidos usando un computador y es una herramienta muy usada por los ingenieros, en el caso de nuestra carrera se utiliza para estudiar el secado, el congelamiento, mezclado, cocción, esterilización etc. 
Algunos ejemplos los podemos encontrar en internet o la literatura sobre la aplicación del CFD y los métodos numéricos en la Ingeniería en alimentos sea la transferencia de calor casos de hornos, secadores, estufas etc. Además de garantizar la inocuidad con la temperatura de diferentes materiales de la industria.

 
**Método de volumen de control CFD**
Consiste en dividir el dominio en un número finito de volúmenes de control (celdas) no superpuestas, en los cuales se aplica los principios de conservación de materia y energía.

$$ \int_{w}^{e}(\frac{d\phi}{dx})dx+\int_{w}^{e}\phi dx=0 $$

Donde e y w son los límites de las paredes del material.
Si aplicamos este concepto y suponemos una variación entre los centros de cada uno de los volúmenes de control establecemos la siguiente ecuación 

$$ \frac{\phi E-\phi P}{\Delta X}-\frac{\phi P-\phi W}{\Delta X}+\bar{\phi }\Delta X=0 $$

$$ a_{P\phi p}=a_{E\phi E}+a_{W\phi W}+b $$

En donde ap, aE y aw son los coeficientes de las variables en cada uno de los nodos implicados en la pared.

Para la deducción de esta ecuación se parte de la ecuación de Fourier 

$$ q=-kA\frac{dT}{dr} $$

$$ q=4\pi k\frac{T_{i}-T_{o}}{(\frac{1}{ri}-\frac{1}{ro})}$$

**Ecuación para el grado crítico de aislamiento**

En este caso se realiza el esquema de resistencias por ejemplo una pared cubierta con un aislamiento la resistencia quedaría con una ecuación para la exterior de la pared del metal (tiende a cero), otra del aislamiento y otra la de convección.
Resistencia del metal que tiende a cero 

$$ \frac{ln(\frac{r_{2}}{r_{1}})}{2\pi lk_{n}} $$
Resistencia para el aislamiento 
$$ \frac{ln(\frac{r_{o}}{r_{i}})}{2\pi lk} $$
Resistencia para la convección
$$ \frac{1}{2\pi l_{ro}k} $$
En donde la ecuación general nos quedaría 
$$ q=-\frac{\Delta T}{\sum_{i=1}^{n}R_{i}}=-\frac{\Delta T}{\sum_{i=1}^{2}R_{i}}=-\frac{\Delta T}{R_{1}+R_{2}} $$
Luego reemplazando las resistencias tenemos que 
$$ q=-\frac{T_{\infty}-T_{i}}{R_{1}-R_{2}}=\frac{T_{i}-T_{\infty }}{\frac{ln(\frac{r_{o}}{r_{i}})}{2\pi lk}+\frac{1}{2\pi l_{ro}k}} $$
Para minimizar esta ecuación empleamos la primera derivada e igualamos a cero (ro representa el radio externo)

$$ \frac{1}{2\pi lr_{o}}\left (\frac{1}{k}-\frac{1}{hr_o}  \right )=0 $$

Condiciones de borde
Se establece la ecuación diferencial para conducción unidimensional de estado estable sin generación de calor que resuelve tanto la temperatura como la velocidad de transferencia de calor, así como el flujo, usando diferentes condiciones de contorno.

$$ \frac{d}{dx}\left (k\frac{dT}{dx}  \right)=0 $$
Como no es una función de x obtenemos

$$ \frac{d^{2}T}{dx^{2}}=0 $$

$$ \frac{dT}{dx}=C1 $$

$$ T(x)=C_{1}x+C_{2} $$
Donde las condicionales son

$$ a) x= 0 T= T_{s1} $$
$$ b) x=L   T=T_{s,2} $$
$$ T_{s1}=C_{2}$$
Luego llegamos a 

$$ T_{s2}=C_{1}L+T_{s1} $$

$$ C_{1}=\frac{T_{s2}-T_{s1}}{L} $$
Y por último llegamos a la ecuación de Fourier 

$$ q=-kA\frac{dT}{dx} $$
En donde k es constante y T la temperatura.
De la ecuación anterior esta debe simplificarse si vamos a analizar una pared plana que está en estado estable, no tiene generación de calor y la conductividad térmica k es constante.
Entonces tenemos que 
 $$ q=-kA\frac{dT}{dx} $$

$$ \frac{\partial }{\partial x} \left(k\frac{\partial T}{\partial x} \right )+ \frac{\partial }{\partial y} \left(k\frac{\partial T}{\partial y} \right )+\frac{\partial }{\partial z} \left(k\frac{\partial T}{\partial z} \right )+\dot{q}=\rho Cp\frac{\partial T}{\partial t} $$

Como es una pared plana no disponemos de los ejes *y* y *z* y solo nos quedaría

$$ \frac{\partial }{\partial x} \left(k\frac{\partial T}{\partial x} \right )=0 $$

Luego integramos dos veces para para obtener una distribución de temperatura para asi regresar a nuestra ecuación original de transferencia de calor  

$$ \frac{d^{2}T}{dx^{2}}=0 $$
Luego tenemos que

$$ T(x)=C_{1}x+C_{2} $$
Estos parámetros se encuentran en los bordes de la pared y se puede establecer las condiciones ya mencionadas anteriormente.
Con esta ecuación se puede derivar y obtener de acuerdo a la geometría del cuerpo o material la ecuación de transferencia de calor basado en el balance general de flujos de entradas, salidas y reacción si ese es el caso.
Para cuerpos rectangulares

Cuerpos cilíndricos

Para el contacto entre las paredes se necesita de armar un esquema de resistencias sea que las paredes se encuentren en serie o en paralelo y se puede resolver con la análoga a la ecuación de Fourier.

$$ {q}''=\frac{\Delta T}{R_{total}} $$
La resistencia total varía de acuerdo a si las paredes se encuentran la serie o en paralelo 
En algunos casos los términos de conducción pueden ser despreciados al igual que las resistencias, por lo cual se llega a una solución mucho más simple. Para un sólido, esta condición puede ser probada vía un numero adimensional llamado número de Biot (Bi). El número de Biot refleja la razón de la resistencia al flujo de calor, interna (conducción) a la externa (convección).

$$ Bi=\frac{Resistencia interna- Conduccion}{Resistencia externa-conveccion}=\frac{\frac{L_{c}}{kA}}{\frac{1}{hA}}=\frac{hL_{c}}{k} $$
Circuitos térmicos
La conductividad térmica k es una propiedad de los materiales que excepto en el caso de los gases a bajas temperaturas, no es posible predecir analíticamente. La información disponible está basada en medidas experimentales. Por otro lado , la conductividad térmica de un material varía con la temperatura, pero en muchas situaciones prácticas, si el sistema tiene una temperatura media, se puede considerar con un valor medio constante, lo que proporciona resultados bastante satisfactorios. 
Se necesitan de dos ecuaciones la Intensidad (I) y el flujo de calor (q)

$$ I=\frac{\Delta V}{Re}$$

$$ q=\frac{\Delta T}{R_{t}}$$
Donde R representa la resistencia 
Conductividad de la resistencia 

$$ R_{cd}=\frac{\Delta T}{q}=\frac{\Delta t}{\left (\frac{kA\Delta T}{L} \right )}=\frac{L}{kA} $$
Covectividad de la resistencia 

$$ R_{cv}=\frac{\Delta T}{q}=\frac{\Delta t}{\left (hA\Delta T \right )}=\frac{1}{hA}$$
