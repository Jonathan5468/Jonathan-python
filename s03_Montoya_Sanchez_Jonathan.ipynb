```python
import numpy as np
import matplotlib.pyplot as plt 

%matplotlib notebook
r1= 102.26/1000
r2= 114.30/1000
T1= 100
T2=80
r= np.linspace(r1, r2, num=10)
T=T2 + ((T1-T2)/np.log(r1/r2))*np.log(r/r2)
Tb= T1 + ((T1-T2)/np.log(r1/r2))*np.log(r/r1)
plt.plot(r,T,'r-',r,Tb,'bs')
plt.show()
```

Linspace con matriz y vectores 

```python
x= np.linspace(-2,2)
y= np.linspace(-2,2) 
xx,yy= np.meshgrid(x,y) 
```

```python
xx,yy
```

```python
def g(x,y):
    return np.cos(x) + np.sin(y)** 2 
```

```python
zz=g(xx,yy)
```

```python
fig, axes = plt.subplots()

axes.contourf(xx, yy, zz, np.linspace(-2, 1), cmap=plt.cm.autumn)
```

```python
fig, axes= plt.subplots()
axes.contour (xx, yy,zz)
```
